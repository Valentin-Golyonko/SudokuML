```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
from numpy import ndarray

from dj_config.settings import BASE_DIR
```

```python
"""
Load board image.
Docs: https://scikit-image.org/docs/stable/user_guide/getting_started.html
"""

difficulty = "easy"
board_number = 81
board_name = f"{difficulty}_{board_number}"

img_path = f"{BASE_DIR}/data_sources/board_img/easy/{board_name}.png"

if not Path(img_path).exists() or not Path(img_path).is_file():
    raise Exception("No file")

img_rgba: ndarray = ski.io.imread(img_path)
```

```python
"""Show image (row data)"""

print(f"{img_rgba.shape = }")

plt.figure()
plt.imshow(img_rgba)
plt.colorbar()
plt.show()
```

```python
"""remove alpha channel"""

img_rgb: ndarray = ski.color.rgba2rgb(img_rgba)
print(f"{img_rgb.shape = }")

plt.figure()
plt.imshow(img_rgb)
plt.colorbar()
plt.show()
```

```python
"""convert to grayscale"""

img_gray = ski.color.rgb2gray(img_rgb)
print(f"{img_gray.shape = }")

plt.figure()
plt.imshow(img_gray)
plt.colorbar()
plt.show()
```

```python
"""invert grayscale"""

board_img = ski.util.invert(img_gray)
print(f"{board_img.shape = }")

plt.figure()
plt.imshow(board_img)
plt.colorbar()
plt.show()
```

```python
"""show processed image (binary)"""

plt.figure()
plt.imshow(board_img, cmap=plt.cm.binary)
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.imshow(board_img[27:34, 102:110], cmap=plt.cm.binary)
plt.colorbar()
```
