```python
from pathlib import Path

import numpy as np
from numpy import ndarray

import matplotlib.pyplot as plt

import skimage as ski
from skimage.exposure import histogram
from skimage.transform import rescale, rotate

from dj_config.settings import BASE_DIR
```

```python
"""
https://scikit-image.org/docs/stable/user_guide/getting_started.html
"""

img_path = f"{BASE_DIR}/media/easy/easy_21.png"

print(f"exists = {Path(img_path).exists()}")
print(f"is_file = {Path(img_path).is_file()}")

img_rgba: ndarray = ski.io.imread(img_path)
```

```python
print(f"{img_rgba.shape = }")

plt.figure()
plt.imshow(img_rgba)
plt.colorbar()
plt.show()
```

```python
img_rgb: ndarray = ski.color.rgba2rgb(img_rgba)
print(f"{img_rgb.shape = }")

plt.figure()
plt.imshow(img_rgb)
plt.colorbar()
plt.show()
```

```python
img_gray = ski.color.rgb2gray(img_rgb)
print(f"{img_gray.shape = }")

plt.figure()
plt.imshow(img_gray)
plt.colorbar()
plt.show()
```

```python
board_img = ski.util.invert(img_gray)
print(f"{board_img.shape = }")

plt.figure()
plt.imshow(board_img)
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.imshow(board_img, cmap=plt.cm.binary)
plt.colorbar()
plt.show()
```

```python
p01 = board_img[14:42, 58:86]
plt.imshow(p01, cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.show()
```
